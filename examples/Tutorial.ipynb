```python
from dtgraph import Neo4jGraph, Rule
```

```python
graph = Neo4jGraph("bolt://localhost:7687", "neo4j", password="password", verbose = True)
```

```python
graph.output_all_nodes()
```

```python
my_query = Rule.from_ascii('''
        MATCH (n:Person)-[:ACTED_IN]->(m:Movie)<-[:ACTED_IN]-(o:Person)
        => 
        (x = (n) : Actor {
            name = n.name,
            born = n.born
        })-[(m) : COLLEAGUE {
            movie = m.title
        }]->(y = (o) : Actor {
            name = o.name,
            born = o.born
        })
    ''')
    my_query.apply_on(graph)
```
